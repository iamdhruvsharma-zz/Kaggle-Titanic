```python
train_data = pd.read_csv("/kaggle/input/titanic/train.csv")
train_data_y = pd.read_csv("/kaggle/input/titanic/train.csv")
train_data_y = train_data_y["Survived"]
test_data = pd.read_csv("/kaggle/input/titanic/test.csv")
test_data_pid = test_data["PassengerId"]
```

```python
train_data = train_data.drop(labels = ["PassengerId", "Name", "Ticket", "Cabin"], axis = 1)
test_data = test_data.drop(labels = ["PassengerId", "Name", "Ticket", "Cabin"], axis = 1)
```

```python
train_data["Embarked"] = train_data["Embarked"].fillna(value = 'S')
test_data["Embarked"] = test_data["Embarked"].fillna(value = 'S')
```

```python
#Handling Age Column
train_data["Age"] = train_data["Age"].fillna(train_data["Age"].median())
test_data["Age"] = test_data["Age"].fillna(test_data["Age"].median())
```

```python
for i in range(0, len(train_data["Age"])):
    age = train_data.iloc[i, 3]
    if age>0.0 and age<=17.0:
        train_data.iloc[i, 3] = "1"
    elif age>17.0 and age<=33.0:
        train_data.iloc[i, 3] = "2"
    elif age>33 and age<=49:
        train_data.iloc[i, 3] = "3"
    elif age>49:
        train_data.iloc[i, 3] = "4"
```

```python
for i in range(0, len(test_data["Age"])):
    age = test_data.iloc[i, 2]
    if age>0.0 and age<=17.0:
        test_data.iloc[i, 2] = "1"
    elif age>17.0 and age<=33.0:
        test_data.iloc[i, 2] = "2"
    elif age>33 and age<=49:
        test_data.iloc[i, 2] = "3"
    elif age>49:
        test_data.iloc[i, 2] = "4"
```
